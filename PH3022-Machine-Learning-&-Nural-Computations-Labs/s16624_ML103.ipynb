Removing rows with missing values is safe when;

    1) Only a few rows are missing values.
    
    2) Missing values are random and not related to specific patterns.
    
    3) Dropping rows does NOT damage the sample size significantly.

Removing rows with missing values is not safe when;

    1) The dataset is small.
    
    2) Many rows contain missing values.
    
    3) Missing values are concentrated in important columns.
    
    4) The missing values are not random.

For this dataset:

Automobile dataset is small(nearly 205 rows) and missing values occur mainly in:

    1) normalized-losses
    
    2) bore
    
    3) stroke
    
    4) horsepower
    
    5) peak-rpm
    
    6) price

Dropping these rows will remove many valuable samples, especially price rows. This is very risky for machine learning.

<b>Q1.h</b>

Fill missing values using:

    Mean for numerical columns
    
    Most frequent (mode) for string/categorical columns

```python
df_mean_mode = df_new.copy()

num_cols = df_mean_mode.select_dtypes(include=['number']).columns
cat_cols = df_mean_mode.select_dtypes(include=['object']).columns

df_mean_mode[num_cols] = df_mean_mode[num_cols].fillna(df_mean_mode[num_cols].mean())

for col in cat_cols:
    df_mean_mode[col] = df_mean_mode[col].fillna(df_mean_mode[col].mode()[0])
```

Fill missing values using:

    Median for numerical columns
    
    Most frequent (mode) for string/categorical columns

```python
df_median_mode = df_new.copy()

num_cols = df_median_mode.select_dtypes(include=['number']).columns
cat_cols = df_median_mode.select_dtypes(include=['object']).columns

df_median_mode[num_cols] = df_median_mode[num_cols].fillna(df_median_mode[num_cols].median())

for col in cat_cols:
    df_median_mode[col] = df_median_mode[col].fillna(df_median_mode[col].mode()[0])
```

<b>Q1.i</b>

One-hot encoding is a technique used to convert categorical (non-numeric) data into a numerical format that machine learning models can understand.

The importance of One-Hot Encoding;

    1) To prevent false numerical relationships.
    
    2) To make categorical data usable by ML models.
    
    3) To help the model learn category-specific patterns.
    
    4) To avoid misleading magnitude-based effects.
    
    5) It works well with sparse, high-dimensional ML algorithms.

<b>Q1.j</b>

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', 2000)

categorical_cols = df_new.select_dtypes(include=['object', 'category']).columns
print("Categorical Columns:", categorical_cols.tolist())

df_ohe = pd.get_dummies(df_new, columns=categorical_cols, drop_first=False)

encoded_cols = []
for cat in categorical_cols:
    encoded_cols.extend([col for col in df_ohe.columns if col.startswith(cat + "_")])

print("\nOne-Hot Encoded Columns:")
print(encoded_cols)

df_ohe[encoded_cols] = df_ohe[encoded_cols].astype(int)

df_ohe[encoded_cols].head()
```
